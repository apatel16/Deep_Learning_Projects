### Computation graph and automatic differentiation

```python
X = autograd.Variable(torch.Tensor([1, 2, 3]), requires_grad=True)

y = autograd.Variable(torch.Tensor([4, 5, 6]), requires_grad=True)

z = X + y
print(z.data)

operation = z.grad_fn
print(operation)
```

```python
s = z.sum()
print(s)
print(s.grad_fn)
```

# Linear Model in Pytorch

```python
import numpy as np
```

### Create data using numpy

```python
x = [i for i in range(20)]
x_train = np.array(x, dtype=np.float32)
x_train = x_train.reshape(-1, 1)
print(x)
print(x_train.shape)

y = [(5*i + 2) for i in x]
y_train = np.array(y, dtype=np.float32)
y_train = y_train.reshape(-1, 1)
print(y)
print(y_train.shape)
```

### Create Model class

```python
class LinearRegressor(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(LinearRegressor, self).__init__()
        self.linear = nn.Linear(input_dim, output_dim)
        
    def forward(self, x):
        out = self.linear(x)
        return out
    
input_dim = 1
output_dim = 1

model = LinearRegressor(input_dim, output_dim)
model
```

### Loss and Optimizer

```python
loss_funtion = nn.MSELoss()

optimizer = torch.optim.SGD(model.parameters(), lr=0.001)
optimizer
```

```python
epoch = 5000
for epoch in range(epoch):
    epoch += 1
    
    inputs = autograd.Variable(torch.from_numpy(x_train))
    real_outputs = autograd.Variable(torch.from_numpy(y_train))
    
    optimizer.zero_grad()
    
    pred_outputs = model(inputs)
    
    loss = loss_funtion(pred_outputs, real_outputs)
    
    #Compute gradients
    loss.backward()
    
    optimizer.step()
    
    print("epoch: {} loss: {} ".format(epoch, loss.item()))
    
```
